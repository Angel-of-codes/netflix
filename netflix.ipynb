```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print('loaded')
```

```python
df = pd.read_csv('Netflix_data.csv')
```

```python
df.info()
```

```python
# re_year = df.release_year.sort_values(ascending = False)
re_year_data = df.release_year[df.release_year>2000].value_counts()
re_year_data
```

```python
plt.figure(figsize= (21,7))
plt.title('Most Released Years')
plt.xlabel('Years')
plt.ylabel('Released')
sns.barplot(re_year_data, order=re_year_data.index.sort_values(ascending=False))
# sns.barplot(df.release_year.sort_values(ascending = False))
```
